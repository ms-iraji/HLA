# Demo: Hybrid-based Learning Automata (HLA)
This notebook demonstrates how to run UFSM, WLA, and HLA methods on a synthetic dataset.

```python
from sklearn.datasets import make_classification
from ufsm import UFSM_filter
from wla import WLA
from hla import HLA

X, y = make_classification(n_samples=200, n_features=100, n_informative=20, random_state=0)
print('Running UFSM...')
ufsm_features = UFSM_filter(X, t=30)
print('Selected features (UFSM):', ufsm_features[:10])

print('Running WLA...')
wla_rank = WLA(X, y, n_clusters=10)
print('Top 5 WLA-ranked features:', wla_rank[:5])

print('Running HLA...')
hla_rank = HLA(X, y, t=30)
print('Top 5 HLA-ranked features:', hla_rank[:5])
```
